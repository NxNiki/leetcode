```python

class Solution:
    def isValidSudoku(self, board: List[List[str]]) -> bool:
        
        def rowValid(i):
            s = set()
            for c in range(9):
                if board[i][c] == '.':
                    continue
                if board[i][c] in s:
                    return False
                s.add(board[i][c])
            return True
                
        def colValid(i):
            s = set()
            for r in range(9):
                if board[r][i] == '.':
                    continue
                if board[r][i] in s:
                    return False
                s.add(board[r][i])
            return True
        
        def squareValid(i, j):
            s = set()
            for r in range(3):
                for c in range(3):
                    if board[i+r][j+c] == '.':
                        continue
                    if board[i+r][j+c] in s:
                        return False
                    s.add(board[i+r][j+c])
            return True
        
        for r in range(9):
            if not rowValid(r):
                print('false1', r)
                return False
        
        for c in range(9):
            if not colValid(c):
                print('false2', c)
                return False
            
        for r in range(0, 9, 3):
            for c in range(0, 9, 3):
                print(r, c)
                if not squareValid(r, c):
                    print('false3')
                    return False
        
        return True
        





board = [[".",".",".",".","5",".",".","1","."],
         [".","4",".","3",".",".",".",".","."],
         [".",".",".",".",".","3",".",".","1"],
         ["8",".",".",".",".",".",".","2","."],
         [".",".","2",".","7",".",".",".","."],
         [".","1","5",".",".",".",".",".","."],
         [".",".",".",".",".","2",".",".","."],
         [".","2",".","9",".",".",".",".","."],
         [".",".","4",".",".",".",".",".","."]]

```
