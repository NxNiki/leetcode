```python
class Solution:
    def jump(self, nums) -> int:

        res = 0
        left = right = 0
        farthest = 0
        
        while left<len(nums):
            print(left) # we get stuck at pos 2 with a 0 value.
            right = left + nums[left]
            res += 1
            for i in range(left, right+1):
                farthest = i + nums[i]
                farthest = max(right, farthest)
            if farthest>=len(nums):
                return res
            
            left = farthest
            
            
s = Solution()
s.jump([2,3,0,1,4])
```
