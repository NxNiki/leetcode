```python
class Solution:
    
    def solveSudoku(self, board) -> None:
        """
        Do not return anything, modify board in-place instead.
        """
        N = 9
        n = 3
        
        # list of set to store current numbers in each col, row, and block.
        cols = [set() for _ in range(N)]
        rows = [set() for _ in range(N)]
        blocks = [set() for _ in range(N)]
        
        # map row, col index to block index:
        block_index = lambda row, col: (row // n) * n + col // n
        
        # whether number d and be placed in row, col on the board:
        def can_place(d, row, col):
            return not (d in rows[row] or 
                        d in cols[col] or 
                        d in blocks[block_index(row, col)])
        
        def place_d(d, row, col):
            board[row][col] = str(d)
            rows[row].add(d)
            cols[col].add(d)
            blocks[block_index(row, col)].add(d)
            
        def remove_d(d, row, col):
            board[row][col] = '.'
            rows[row].remove(d)
            cols[col].remove(d)
            blocks[block_index(row, col)].remove(d)
            
        def place_next(row, col):
            if row == N - 1 and col == N - 1:
                nonlocal solved
                solved = True
            else:
                if col == N - 1:
                    backtrack(row + 1, 0)
                else:
                    backtrack(row, col + 1)
                    
        def backtrack(row, col):
            
            if board[row][col] == '.':
                for d in range(1, 10):
                    if can_place(d, row, col):
#                         print(d, row, col)
                        place_d(d, row, col)
                        place_next(row, col)
                        if not solved:
                            remove_d(d, row, col)
                        
            else:
                place_next(row, col)
                
        
        for i in range(N):
            for j in range(N):
                if board[i][j] != '.':
                    place_d(int(board[i][j]), i, j)
                
        solved = False
        backtrack(0, 0)

board = [["5","3",".",".","7",".",".",".","."],
         ["6",".",".","1","9","5",".",".","."],
         [".","9","8",".",".",".",".","6","."],
         ["8",".",".",".","6",".",".",".","3"],
         ["4",".",".","8",".","3",".",".","1"],
         ["7",".",".",".","2",".",".",".","6"],
         [".","6",".",".",".",".","2","8","."],
         [".",".",".","4","1","9",".",".","5"],
         [".",".",".",".","8",".",".","7","9"]]

sol = Solution()
sol.solveSudoku(board)
board
```

```python
class Solution:
    def solveSudoku(self, board: List[List[str]]) -> None:
        """
        Do not return anything, modify board in-place instead.
        """
        
        n = len(board)

        rows, cols, boxes = collections.defaultdict(set), collections.defaultdict(set), collections.defaultdict(set)

        def is_valid(r, c, v):
            return v not in rows[r] and v not in cols[c] and v not in boxes[(r//3, c//3)]
        
        def place_d(v, r, c):
            board[r][c] = str(v)
            rows[r].add(v)
            cols[c].add(v)
            boxes[(r//3, c//3)].add(v)
            
        def remove_d(v, r, c):
            board[r][c] = '.'
            rows[r].remove(v)
            cols[c].remove(v)
            boxes[(r//3, c//3)].remove(v)

        def backtrack(r, c):
            if r == n - 1 and c == n:
                return True
            elif c == n:
                c = 0
                r += 1

            # current grid has been filled
            if board[r][c] != '.':
                return backtrack(r, c + 1)

            for v in range(1, n + 1):
                if not is_valid(r, c, v):
                    continue

                place_d(v, r, c)

                if backtrack(r, c + 1):
                    return True

                # backtrack
                remove_d(v, r, c)

            return False
        
        for r in range(n):
            for c in range(n):
                if board[r][c] != '.':
                    v = int(board[r][c])
                    place_d(v, r, c)
                
        backtrack(0, 0)
        
```
